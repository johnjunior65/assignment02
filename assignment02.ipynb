```python
import numpy as np
import matplotlib.pyplot as plt
```

# define my function : $f(x) = \sin(x) \bullet x$

```python
def myFunction(x):
    f=np.sin(x)*x
    return f
```

# define the derivative of my function : $f'(x) = \sin(x) + cos(x) \bullet x$

```python
def myDerivativeFunction(x):
    Df=np.sin(x)+np.cos(x)*x
    return Df
```

# define the domain of the function : $x = [-10 : 0.1 : 10]$

```python
x=np.arange(-10,10,0.1)
```

# pick 3 points in the domain $x_1 = -4, x_2 = 0, x_3 = 6$

```python
x1=-4
x2=0
x3=6
```

# plot the graph of the defined function

```python
f=myFunction(x)
Df=myDerivativeFunction(x)
plt.figure(1)
plt.plot(x, f, 'b', label='function')
plt.plot(x, Df, 'r', label='derivative')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# taylor approximation for given function
### taylor는 테일러 근사 다항식을 정의한 것입니다.
### 테일러 근사 다항식의 n값을 조절하여 다항식이 원하는 식에 근사해가는 과정을 볼 수 있게 하였습니다.

```python
def taylor(x, n):
    a = x**2
    t = a
    for i in range(1, n):
        a *= -1 * x**2 / ((2 * i) * (2 * i + 1))
        t += a
    return t

ty = np.vectorize(taylor, excluded=['n'])

n5 = ty(x, 5)
n10 = ty(x, 10)
n100 = ty(x, 100)
n1000 = ty(x, 1000)
n10000 = ty(x, 10000)

plt.plot(x, f, label='sin(x)*x')
plt.plot(x, n5, label='n = 5')
plt.plot(x, n10, label='n = 10')
plt.plot(x, n100, label='n = 100')
plt.plot(x, n1000, label='n = 1000')
plt.plot(x, n10000, label='n = 10000')

plt.ylim([-10, 10])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
